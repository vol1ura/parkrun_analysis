```python
import numpy as np
import pandas as pd
import random
import requests
import time
from lxml.html import parse
from tqdm.notebook import tqdm

pd.set_option('display.max_rows', None)
```

```python
parkrun = 'kuzminki'
```

```python
url = f'https://www.parkrun.ru/{parkrun}/results/agecategoryrecords/'
headers = {
    'Host': 'www.parkrun.ru',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:88.0) Gecko/20100101 Firefox/88.0',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Language': 'ru-RU,ru;q=0.8,en-US;q=0.5,en;q=0.3',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1',
    'Sec-GPC': '1',
    'TE': 'Trailers'
    }
```

```python
result = requests.get(url, headers=headers, stream=True)
result.raw.decode_content = True
tree = parse(result.raw)
```

```python
categories = tree.xpath('//tbody/tr/td[1]/a/@href')
```

```python
dfs = []
```

```python
for category in tqdm(categories):
    time.sleep(3+5*random.random())  # ЕСЛИ завершилось с ошибкой, то можно поставить циферку побольше и попробовать через некоторое время снова
    url = f'https://www.parkrun.ru/{parkrun}/results/agecategoryrecords/{category}'
    age_results = requests.get(url, headers=headers)
    try:
      df = pd.read_html(age_results.text)[0]
    except:
      print('ОШИБКА - операция завершилась досрочно. Паркран временно заблокировал.')
      break
    df['Age'] = category.split('=')[1]
    dfs.append(df)
```

```python
df_all = pd.concat(dfs)
df_all['Клуб'].fillna('', inplace=True)
```

```python
df_all = df_all.groupby(by=['Участник']).agg({'Число пробежек': np.sum, 'Лучшее время': ['min'], 'Клуб': ['max'], 'Age': ['max']}) 
df_all.columns = ['Число пробежек', 'Лучшее время', 'Клуб', 'Age']
df_all = df_all.sort_values(by='Число пробежек', ascending=False)
df_all.reset_index()
```
